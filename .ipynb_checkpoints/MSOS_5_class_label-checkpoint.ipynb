```python
Features=da.iloc[:,2:100]
labels=da.iloc[:,0]
labels.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(Features,labels, test_size=.20, random_state=40) 
np.unique(y_test,return_counts=True)
```

```python
cv=5
from sklearn.model_selection import GridSearchCV
parameters = {'kernel':('linear', 'rbf','poly'), 'C':[1, 10,100,1000],'gamma':[0.01,.002,0.003]}
svc = svm.SVC()
clf = GridSearchCV(svc, parameters,cv=cv)
clf.fit(X_train, y_train)
```

```python
pred=clf.best_estimator_.predict(X_test)
#clf.score(X_test, y_test)  
print pred
print y_test
```

```python
acc=accuracy_score(clf.best_estimator_.predict(X_test),y_test)
print acc
```

```python
conf_mat=confusion_matrix(y_test,pred)
conf_mat
```
